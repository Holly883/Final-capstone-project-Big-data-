```python
import pandas as pd

# Step 1: Load raw dataset
df = pd.read_csv("tb_data.csv.csv")

# Step 2: Keep only useful columns
df_clean = df[["COUNTRY", "YEAR", "VALUE"]].copy()

# Step 3: Rename columns for clarity
df_clean.rename(columns={
    "COUNTRY": "Country",
    "YEAR": "Year",
    "VALUE": "TB_Cases"
}, inplace=True)

# Step 4: Drop missing values
df_clean.dropna(subset=["Country", "Year", "TB_Cases"], inplace=True)

# Step 5: Convert data types
df_clean["Year"] = df_clean["Year"].astype(int)
df_clean["TB_Cases"] = pd.to_numeric(df_clean["TB_Cases"], errors="coerce")

# Step 6: Drop rows that couldnâ€™t convert to numeric
df_clean.dropna(subset=["TB_Cases"], inplace=True)

# Step 7: Remove countries with 0 TB cases in all years (optional but cleaner)
country_sums = df_clean.groupby("Country")["TB_Cases"].sum()
valid_countries = country_sums[country_sums > 0].index
df_clean = df_clean[df_clean["Country"].isin(valid_countries)]

# Step 8: Reset index (neatness)
df_clean.reset_index(drop=True, inplace=True)

# Step 9: Export cleaned data for use in Power BI & GitHub
df_clean.to_csv("cleaned_tb_trends.csv", index=False)

# Show preview
df_clean.head(10)
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import seaborn as sns
import matplotlib.pyplot as plt

# Group data: average TB cases per country
df_avg_tb = df.groupby('COUNTRY')['VALUE'].mean().reset_index()
df_avg_tb.columns = ['COUNTRY', 'AVG_TB_CASES']

# Standardize data for clustering
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_avg_tb[['AVG_TB_CASES']])

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df_avg_tb['CLUSTER'] = kmeans.fit_predict(scaled_data)

# Plot clustering results
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_avg_tb.index, y=df_avg_tb['AVG_TB_CASES'], hue=df_avg_tb['CLUSTER'], palette='viridis')
plt.title("Clustering Countries by Average TB Cases")
plt.xlabel("Country Index")
plt.ylabel("Average TB Cases")
plt.show()

# Evaluate clustering using silhouette score
score = silhouette_score(scaled_data, df_avg_tb['CLUSTER'])
print(f"Silhouette Score: {score:.3f}")

# Print top country per cluster
for cluster in sorted(df_avg_tb['CLUSTER'].unique()):
    top_country = df_avg_tb[df_avg_tb['CLUSTER'] == cluster].sort_values('AVG_TB_CASES', ascending=False).iloc[0]
    print(f"Top TB Country in Cluster {cluster}: {top_country['COUNTRY']} with {top_country['AVG_TB_CASES']:.0f} cases")
```

```python
def cluster_summary(df, cluster_col='CLUSTER', cases_col='AVG_TB_CASES', country_col='COUNTRY'):
    """
    Summarizes cluster statistics:
    - Number of countries in each cluster
    - Average TB cases in each cluster
    - Top country (highest TB cases) in each cluster
    
    Parameters:
    - df: DataFrame with clusters
    - cluster_col: column name for cluster labels
    - cases_col: column name for TB cases
    - country_col: column name for country names
    
    Returns:
    - summary_df: DataFrame with the summary
    """
    summary_list = []

    for cluster in sorted(df[cluster_col].unique()):
        cluster_data = df[df[cluster_col] == cluster]
        num_countries = cluster_data.shape[0]
        avg_cases = cluster_data[cases_col].mean()
        top_country = cluster_data.sort_values(cases_col, ascending=False).iloc[0][country_col]
        top_cases = cluster_data[cases_col].max()
        
        summary_list.append({
            'Cluster': cluster,
            'Num_Countries': num_countries,
            'Average_TB_Cases': avg_cases,
            'Top_Country': top_country,
            'Top_Country_Cases': top_cases
        })
    
    summary_df = pd.DataFrame(summary_list)
    return summary_df
```

```python
summary = cluster_summary(df_avg_tb)
print(summary)
```

# TB Case Trends Across Different Countries

This project analyzes tuberculosis (TB) case trends across countries using WHO data from 2023. The goal is to identify high burden countries and find patterns using data analysis and clustering.

---

## 1. Data Cleaning

We loaded the raw dataset and selected relevant columns. Missing values were handled, and column names standardized for easier analysis.

---

## 2. Exploratory Data Analysis (EDA)

We summarized total TB cases per country and visualized the top 10 countries with highest incidence. This helped identify countries with the greatest TB burden.

---

## 3. Machine Learning: KMeans Clustering

We applied KMeans clustering on average TB cases per country to group countries with similar TB burdens. Three clusters were identified, with a high silhouette score of 0.931 indicating good cluster separation.

### Innovation: Custom Cluster Summary Function

A custom function was created to summarize each cluster, including number of countries, average cases, and the top country by cases.

---

## 4. Results and Insights

- Cluster 0 top country: Democratic Republic of the Congo (~21,000 cases)  
- Cluster 1 top country: India (~175,000 cases)  
- Cluster 2 top country: Indonesia (~68,000 cases)

These clusters can help prioritize TB control efforts by grouping countries with similar epidemiological profiles.
